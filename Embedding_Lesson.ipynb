```python
import torch
import torch.nn as nn

# Assuming vocabulary size of 1000 and embedding dimension of 50
embedding_layer = nn.Embedding(
    num_embeddings=1000, embedding_dim=50)
```

```python
# Example input with indices representing words (or other categorical values)
input_data = torch.LongTensor(
    [1, 5, 7, 42])  # Example word indices
embedded_output = embedding_layer(input_data)
```

```python
print(embedded_output)
```
